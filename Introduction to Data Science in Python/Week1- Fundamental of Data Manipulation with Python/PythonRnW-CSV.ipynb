```python
import csv
%precision 2
```

```python
with open('/mpg.csv') as csvfile:
  mpg = list(csv.DictReader(csvfile))

mpg[0:3]
```

```python
len(mpg)
```

```python
mpg[0].keys()
```

```python
#average city mpg
sum(float(d['cty']) for d in mpg) / len(mpg)
```

```python
#average city mpg alternative
city_list = [float(d['cty']) for d in mpg]
sum(city_list) / len(mpg)
```

```python
#average highway mpg
sum(float(d['hwy']) for d in mpg) / len(mpg)
```

```python
#average highway mpg alternative
mpg_list = [float(d['hwy']) for d in mpg]
sum(mpg_list) / len(mpg)
```

```python
cylinders = set(d['cyl'] for d in mpg)
cylinders
```

```python
CtyMpgByCyl = []
for c in cylinders:
  summpg = 0
  cyltypecount = 0
  for d in mpg:
    if d['cyl'] == c:
      summpg += float(d['cty'])
      cyltypecount += 1
  CtyMpgByCyl.append((c, summpg / cyltypecount)) 
```

```python
CtyMpgByCyl
```

```python
CtyMpgByCyl[0]
```

```python
CtyMpgByCyl.sort(key=lambda x: x[0])
CtyMpgByCyl
```

```python
CtyMpgByCyl.sort(key=lambda x: x[1])
CtyMpgByCyl
```

Key paramater untuk menentukan fungsi yang dipanggil pada setiap elemen (dalam hal ini elemen list of set/tupple)

lambda x: x[0]
return indeks pertama dari setiap elemen dari list of set/tupple

```python
# average of highway mpg accross group by different vehicle classes
vehicle_class = set(d['class'] for d in mpg)
vehicle_class 
```

```python
HwyMpgByClass = []
for c in vehicle_class:
  classTypeCount = 0
  sumHighwayMpg = 0
  for d in mpg:
    if d['class'] == c:
      sumHighwayMpg += float(d['hwy'])
      classTypeCount +=1
  HwyMpgByClass.append((c, sumHighwayMpg / classTypeCount))
HwyMpgByClass
```

```python
HwyMpgByClass.sort(key=lambda x: x[1])
HwyMpgByClass
```
