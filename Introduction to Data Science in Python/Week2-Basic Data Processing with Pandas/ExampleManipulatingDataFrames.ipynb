In this lecture we're going to walk through a basic data cleaning process and explore a few more pandas API functions

```python
# lets start by bringing in pandas
import pandas as pd 
# and load our dataset. we're going to be cleaning the list of presidents in the US from wikipedia
df = pd.read_csv('/content/presidents.csv')
# and lets just take a look at some of the data
df.head()
```

```python
# ok, we have some presidents, some dates, we see a bunch of footnotes in the "Born" column which
# might cause issues. lets start with cleaning up that name into firstname and lastname. we're going to
# tackle this with a regex. so we want to create two new columns and apply a regex to the projection of
# the "president" column.

# here's one solution, we could make a copy of the President column
df["First"] = df['President']
df.head()
```

```python
# then we can call replace() and just have a pattern that matches the last name and set it to an empty string
df["First"] = df["First"].replace("[ ].*", "", regex=True)
# now lets take a look
df.head()
```

```python
# thats works, but its kind of gross. and its slow, since we had to make a full copy of a column then
# go through and update strings. there are a few other ways we can deal with this. the most general one
# is use appply() function. lets drop the column we made first
del(df['First'])
df.head()
```

```python
# the apply() function on a dataframe will take some arbritrary function you have written and apply it
# to either a Series (a single column) or DataFrame across all rows or columns. lets write a function
# which just split a string into two pieces using a single row of data
def splitname(row):
  # the row is a single Series object which is a single row indexed by column values
  # lets extract the firstname and create a new entry in the series
  row['First'] = row['President'].split(" ")[0]
  # lets do the same with the lastname
  row['Last'] = row['President'].split(" ")[-1]
  # now we just return the row and the pandas .apply() will take of merging them back into a DataFrame
  return row

# now if we apply this to the dataframe indicating we want to apply it across columns
df = df.apply(splitname, axis='columns')
# axis='columns' equal to axis=1
df.head()
```

```python
del(df['First'])
del(df['Last'])
df.head(20)
```

```python
pattern = "(^[\w]+)(?:.* )([\w]+$)"
df["President"].str.extract(pattern).head(20)
```

```python
# so lets looks pretty nice, other than the column names. but if we name the groups we get named columns out
pattern = "(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"

# now call extract
names = df["President"].str.extract(pattern)
names
```

```python
# and we can just copy these into our main dataframe
df["First"] = names['First']
df["Last"] = names['Last']
df.head()
```

```python
# its worth looking at the pandas str module for other functions which have been written specifically
# to clean up strings in DataFrames, and we can find that in the docs in the Working with Text
# section: https://pandas.pydata.org/pandas-docs/stable/user_guide/text.html
```

```python
# now lets move on to clean up that Born column. first, lets get rid of anything that isnt in the
# pattern of Month day and Year.
df["Born"] = df["Born"].str.extract("([\w]{3} [\w]{1,2}, [\w]{4})")
df["Born"].head()
```

```python
# so, that cleans up the date format. But we're going to foreshadow  something else here - the type
# of this column is object, and we know thats waht pandas uses when it is dealing with string. but pandas
# actually has really interesting date/time features - in fact, thats one of the reasons Wes McKinney put his
# efforts into the library, to deal with financial transactions. so we'll update this column to the datetime data type
df["Born"] = pd.to_datetime(df["Born"])
df["Born"].head()
```

```python
# this would make subsequent processing on the dataframe around dates, such as getting every President
# who was born in a given time span, much easier.
```

The str module has a number of important functions for cleaning pandas dataframes. We dont have to use these, especially if we dont need high performance data cleaning because the dataset is small (we can use apply() funciton). But the str functions are incredibly usefull and build on existing knowledge of regex, and because they are vectorized they are efficient to use as well.
