In this course, we'll be largely using smaller or moderate-sized datasets. A common workflow is to read the dataset in, usually from some external file, then begin to clean and manipulate the dataset for analysis. In this lecture we're going to demonstrate how we can load data from a comma separated file into a DataFrame

```python
# lets jump right in and talk about comma separated values (CSV) files. we've undoubtedly used these
# - any spreadsheet software like excel or google sheets can save our output in CSV format. its pretty
# loose as a format, and incredibly lightweight. and totaly ubiquitous

# now, we're going to make a quick aside because its convenient here. The Jupyter notebooks use ipython as
# the kernel underneath, which provides convenient ways to integrate lower level shell commnads, which are
# programs run in the underlying operating system. If you're not familiar with the shell dont wory too much
# about this, but if you are, this is super handy for integration of your data science workflows. i want to
# use one shell command here called "cat", for "concatenate", which just outputs the contents of a file. in
# ipython if we prepend the line with the an exclamation mark it will execute the remainder of the line as a shell
# command. so lets look at the content of a CSV file
!cat /content/Admission_Predict.csv
```

```python
# we see from the output that there is a list of columns, and the column identifiers are listed as string on
# the first file of the file. then we have rows of data, all columns separated by commas. now, there are lots
# of oddities with the CSV file format, and there is no one agreed upon specification. so you should be
# prepared to do a bit of work when you pull down CSV files to explore. but this lecture isnt focused on CSV
# files, and is more about pandas DataFrames. so lets jump into that.

# lets bring in pandas to work with
import pandas as pd

# Pandas mades it easy to turn a CSV into a DataFrame, we just call read_csv()
df = pd.read_csv('/content/Admission_Predict.csv')

# and lets look at the first few rows
df.head()
```

```python
# we notice that by default index starts with 0 while the students serial number starts from 1.
# if you jump back to the CSV output you'll deduce that pandas has create a new index. Instead,
# we can set the serial no. as the index if we want to by using the index_col.
df = pd.read_csv('/content/Admission_Predict.csv', index_col=0)
df.head()
```

```python
# notice that we have two columns "SOP" and "LOR" and probably not everyone knows what they mean
# so lets change our column names to make it more clear. In Pandas, we can use rename() function
# it takes a parameter called columns, and we need to pass into a dictionary which the keys are the
# old column name and the value is the corresponding new column name

new_df = df.rename(columns={'GRE Score':'GRE Score', 'TOEFL Score':'TOEFL Score',
                            'University Rating':'University Rating', 'SOP':'Statement of Purpose',
                            'LOR':'Letter of Recomendation', 'CGPA':'CGPA', 'Research':'Research',
                            'Chance of Admit':'Chance of Admit'})
new_df.head()
```

```python
# from the output, we can see that only "SOP" is changed but not "LOR". Why is that? lets
# investigate this a bit. first we need to make sure we got all the column names correct. we can use
# the columns attribute of dataframe to get a list
new_df.columns
```

```python
# if we look at the output closely, we can see that there is actually a space right after "LOR" and a
# space right after "Chance of Admit". sneaky, huh? so this is why our rename dictionary does not work for LOR,
# because the key we used was just three characters, instead of "LOR "

# there are a couple of ways we could adress this. one way would be to change a column by including the space
# in the name
new_df = new_df.rename(columns={'LOR ': 'Letter of Recomendation'})
new_df.head()
```

```python
# so that works well, but its a bit fragile. what if that was a tab instead of a space? or two spaces?
# another way is to create some function that does the cleaning and then tell renamed to apply that function
# across all of the data. python comes with a handy string fucntion to strip white space called "strip()".
# when we pass this in to rename we pass the function as the mapper parameter, and then indicate whether the
# axis should be columns or index (row labels)
new_df = new_df.rename(mapper=str.strip, axis='columns')
new_df.head()
```

```python
# now we've got it - both SOP and LOR have been renamed and Chance of Admit has been trimmed up. remember
# though that the rename function isnt modifying the DataFrame. in this case, df is the same as it always
# was, ther's just a copy in new_df with the changed names
df.columns
```

```python
# we can also use the df.columns attribute by assigning to it a list of columns names which will directly
# rename the columns. this will directly modify the original dataframe and is very efficient especially when
# you have a lot of columns and you only want to change a few. this technique is also not affected by subtle
# errors in the column names, a problem that we just encountered. with a list, you can use the list index to
# change a certain value or use list comprehension to change all of the values

# as an example, lets change all of the column names to lower case. first we need to get our list
cols = list(df.columns)
cols
```

```python
# then a little list comprehension
cols = [x.lower().strip() for x in cols]
cols
```

```python
# then we just overwrite what is already in the .columns attribute
df.columns = cols
df.head()
```

```python
# now we will change "sop" to "statement of purpose" and "lor" to "leter of recomendation"
cols[3] = "statement of purpose"
cols[4] = "leter of recomendation"
cols
```

```python
# then we just overwrite what is already in the .columns attribute
df.columns = cols
df.head()
```
